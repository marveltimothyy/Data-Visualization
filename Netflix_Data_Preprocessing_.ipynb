## Preprocessing

###NaN Handling

```python
#NaN Checking
df.isnull().sum()
```

```python
#Drop NaN
df = df.dropna()
df.isnull().sum()
```

```python
df.country.empty
```

```python
df.to_csv('/content/drive/MyDrive/Marvel/Project/Data Visualization/NotNull_Netflix_Dataset.csv')
```

### Split Multivalue

```python
#Set multivalue into split list
df.country = df.country.str.split(',')
df
```

```python
#spread listed value into different rows 
result = df.explode('country')
result
```

Extra Null handling after split processed

```python
(result.country=='').sum()
```

```python
result = result[result.country != '']
(result.country=='').sum()
```

```python
result
```

```python
result.to_csv('/content/drive/MyDrive/Marvel/Project/Data Visualization/Processed_Netflix_Dataset.csv')
```

### One Hot Encoding Process to split movie and tv show in type column

```python
#using OHE function from Sklearn.preprocessing for action 
OHE = OneHotEncoder()
encoder_df = pd.DataFrame(OHE.fit_transform(result[['type']]).toarray())

final_df = result.join(encoder_df)
final_df
```

```python
#take current list of columns
list_columns = final_df.columns
list_columns
```

```python
#manual fit the new columns name 
final_df.columns = ['show_id', 'type', 'title', 'cast', 'date_added',
       'release_year', 'rating', 'duration','description','director','country','listed','movie', 'tv_show']
```

```python
final_df
```

```python
#Save data into csv
final_df.to_csv('/content/drive/MyDrive/Marvel/Project/Data Visualization/Processed_Type_Netflix.csv')
```

##Country Map
Using processed data to be customized for bubble map needs

```python
df_country = pd.read_csv('/content/drive/MyDrive/Marvel/Project/Data Visualization/Processed_Netflix_Dataset.csv')
df_country.head(5)
```

```python
#get number of content for each country
country_count = df_country.country.value_counts().tolist()
print(country_count)
```

```python
#get country list
country_list = df_country.country.unique().tolist()
```

```python
#concat country and it number of content into a list
country = list(zip(country_list, country_count))
```

```python
#set list into Data Frame with named columns
country = pd.DataFrame(country, columns=['country', 'total'])
```

```python
#Save Data 
country.to_csv('/content/drive/MyDrive/Marvel/Project/Data Visualization/Country_Content_Total.csv')
```
